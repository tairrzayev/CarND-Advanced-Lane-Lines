## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

```python
# Helper function to plot 2 images side-by side.
def plot_images(image1, image1_title, image2, image2_title):
    plt.figure(figsize=(24,9))
    plt.subplot(1, 2, 1)

    plt.title(image1_title)
    plt.imshow(image1)

    plt.subplot(1, 2, 2)
    plt.title(image2_title)
    plt.imshow(image2)

    plt.show()
```

## Thresholding

```python
# Helper function to read specific frame of a video
def read_video_frame(file_name, frame_number):
    vidcap = cv2.VideoCapture(file_name)
    vidcap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    success, frame = vidcap.read()
    if not success:
        raise Exception("Failed to read video frame")
    
    return cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

### Thresholding (project video)

```python
frame = read_video_frame('project_video.mp4', 30)
frame_undistorted = lane_detector.undistort(frame)
frame_binary = lane_detector.binary_image(frame_undistorted)

plt.figure(figsize=(24,9))
plt.subplot(1, 2, 1)
plt.title('Original frame #30 from the project video')
plt.imshow(frame_undistorted)

plt.subplot(1, 2, 2)
plt.imshow(frame_binary, cmap='gray')
plt.title('Thresholded frame #30 from the project video')

plt.show()
```

## Perspective transform

```python
image = cv2.imread('./test_images/straight_lines1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = lane_detector.undistort(image)

warped_image = lane_detector.warp(image)#
warped = lane_detector.warp(image)

src = np.array([lane_detector.perspective_transform_src], np.int32)
image = cv2.polylines(image, src, True, (255, 0, 0), 4)

dst = np.array([lane_detector.perspective_transform_dst], np.int32)
warped_image = cv2.polylines(warped_image, dst, True, (255, 0, 0), 4)


plot_images(image, 'straight_lines1.jpg', warped_image, 'straight_lines1.jpg warped')
```

## Fitting the polynomial

```python
lane_detector.reset()
lane_detector.set_debug_enabled(True)

# Fitting the very first video frame
frame_0 = read_video_frame('project_video.mp4', 0)
frame_0_undistorted = lane_detector.undistort(frame_0)
frame_0_original = frame_0_undistorted
frame_0_binary = lane_detector.binary_image(frame_0_undistorted)
frame_0_warped = lane_detector.warp(frame_0_binary)
ploty, left_fitx, right_fitx, debug_output = lane_detector.fit_lane_lines(frame_0_warped)

plt.figure(figsize=(24,9))
plt.subplot(1, 2, 1)
plt.title('Original frame #0 from the project video')
plt.imshow(frame_0_undistorted)

plt.subplot(1, 2, 2)
plt.imshow(debug_output)
plt.plot(left_fitx, ploty, color='yellow')
plt.plot(right_fitx, ploty, color='yellow')
plt.title('Warped frame #0 from the project video with the polynomial fitted')

plt.show()

# Fitting the 2nd video frame, now searching around the previous fit instead of using the sliding window
frame_1 = read_video_frame('project_video.mp4', 1)
frame_1_undistorted = lane_detector.undistort(frame_1)
frame_1_original = frame_1_undistorted
frame_1_binary = lane_detector.binary_image(frame_1_undistorted)
frame_1_warped = lane_detector.warp(frame_1_binary)
ploty, left_fitx, right_fitx, debug_output = lane_detector.fit_lane_lines(frame_1_warped)

plt.figure(figsize=(24,9))
plt.subplot(1, 2, 1)
plt.title('Original frame #1 from the project video')
plt.imshow(frame_1_undistorted)

plt.subplot(1, 2, 2)
plt.imshow(debug_output)
plt.plot(left_fitx, ploty, color='yellow')
plt.plot(right_fitx, ploty, color='yellow')
plt.title('Warped frame #1 from the project video with the polynomial fitted')

plt.show()
```

## Plot the detected lane area back onto the original image

```python
lane_detector.reset()
frame_0 = read_video_frame('project_video.mp4', 0)
frame_0_undistorted = lane_detector.undistort(frame_0)
frame_0_original = frame_0_undistorted
frame_0_binary = lane_detector.binary_image(frame_0_undistorted)
frame_0_warped = lane_detector.warp(frame_0_binary)
ploty, left_fitx, right_fitx, _ = lane_detector.fit_lane_lines(frame_0_warped)
image = lane_detector.unwarp_binary_and_add_to_image(frame_0_original, frame_0_warped, ploty, left_fitx, right_fitx)

plot_images(frame_0_undistorted, 'Original frame #0 from the project video', image, 'Frame #0 from the project video with lane area plotted back onto it')
```

## Define the pipeline

```python
def pipeline(image):
    image = lane_detector.undistort(image)
    image_original = image
    image = lane_detector.binary_image(image)
    image = lane_detector.warp(image)
 
    ploty, left_fitx, right_fitx, _ = lane_detector.fit_lane_lines(image)
    image = lane_detector.unwarp_binary_and_add_to_image(image_original, image, ploty, left_fitx, right_fitx)
    image = lane_detector.draw_lane_data(image, ploty)
    return image
```

## Run the pipeline on project videos

### Project video

```python
lane_detector.reset()
video = VideoFileClip("project_video.mp4")
output = 'project_video_processed.mp4'
white_clip = video.fl_image(pipeline)
%time white_clip.write_videofile(output, audio=False)
```

### Challenge video

```python
lane_detector.reset()
video = VideoFileClip("challenge_video.mp4")
output = 'challenge_video_processed.mp4'
white_clip = video.fl_image(pipeline)
%time white_clip.write_videofile(output, audio=False)
```
